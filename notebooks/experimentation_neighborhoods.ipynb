```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
```

```python
#This method plots the t-SNE plots of the neighborhood of the given word. It first
#performs PCA to reduce the dimensionality of the word vectors and then fits a TSNE model.
def plot_tsne(keyword, m, perplex=1, pca_dims=None, topn=10):
    neighborhood = m.get_nearest_neighbors(keyword, k=topn)
    vecs = [m.get_word_vector(keyword)]
    for _, neighbor in neighborhood:
        vecs.append(m.get_word_vector(neighbor))
    vecs = np.array(vecs)
    reduced = PCA(n_components=pca_dims).fit_transform(vecs)
    projected = TSNE(n_components=2,
                     learning_rate='auto',
                     init='random',
                     perplexity=perplex).fit_transform(reduced)
    df = pd.DataFrame({'x': [x for x in projected[:, 0]],
                       'y': [y for y in projected[:, 1]],
                       'words': [(1.0, keyword)] + neighborhood,
                       'color': ["green"] + ["blue"]*(len(projected)-1)})
    
    fig, _ = plt.subplots()
    fig.set_size_inches(9, 9)
    # plot results
    p1 = sns.regplot(data=df,
                     x="x",
                     y="y",
                     fit_reg=False,
                     marker="o",
                     scatter_kws={'s': 40,
                                  'facecolors': df["color"]
                                 }
                    )
    #Add annotations
    for line in range(0, df.shape[0]):
        p1.text(df["x"][line],
                 df['y'][line],
                 '  ' + df["words"][line][1],
                 horizontalalignment='left',
                 verticalalignment='bottom', size='medium',
                 color=df['color'][line],
                 weight='normal'
                ).set_size(15)
    
    
    plt.xlim(projected[:, 0].min()-50, projected[:, 0].max()+50)
    plt.ylim(projected[:, 1].min()-50, projected[:, 1].max()+50)
            
    plt.title('t-SNE Visualisierung für "{}"'.format(keyword))
```

```python
plot_tsne("asyl", model, perplex=10, topn=10)
```

```python
plot_tsne("chemnitz", model, perplex=10, topn=10)
```
